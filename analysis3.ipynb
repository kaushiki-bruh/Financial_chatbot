This report analyzes the financial performance of three major technology companies—Microsoft, Tesla, and Apple—over the last three fiscal years (2022–2024). The goal is to identify trends and growth patterns that can inform strategic decision-making or investment evaluation.

```python
import pandas as pd
df = pd.read_csv('E:\Cleaned_Financials_Microsoft_Tesla_Apple_2022_2024.csv')
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
df
```

using matplotlib for analyzing revenue

```python
import matplotlib.pyplot as plt

for company in df["Company"].unique():
    subset = df[df["Company"] == company]
    plt.plot(subset["Fiscal Year"], subset["Total Revenue"], label=company)

plt.title("Total Revenue Over Time")
plt.xlabel("Fiscal Year")
plt.ylabel("Revenue")
plt.legend()
plt.show()
```

Revenue insights:
Microsoft's revenue consistently increased each year, indicating stable growth and strong business performance over this period.
Tesla experienced stagnation for one year, likely due to market or operational challenges, but rebounded in 2024, suggesting recovery or expansion efforts.
Despite leading in total revenue, Apple shows a downward trend, which may point to saturation in key markets, increased competition, or lower sales in specific product categories.

```python
avg_by_company = df.groupby("Company").mean(numeric_only=True)
print("Average Financial Metrics by Company:")
print(avg_by_company)

sum_by_year = df.groupby("Fiscal Year").sum(numeric_only=True)
print("\nTotal Financial Metrics by Year (All Companies Combined):")
print(sum_by_year)

df["Asset-to-Liability Ratio"] = df["Total Assets"] / df["Total Liabilities"]

asset_liability_ratio = df.groupby("Company")["Asset-to-Liability Ratio"].mean()
print("\nAverage Asset-to-Liability Ratio per Company:")
print(asset_liability_ratio)
```
